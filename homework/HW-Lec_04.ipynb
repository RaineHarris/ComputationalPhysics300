```python
#MandleBrot Set Attempt
def MandleBrot(niter, thresh, nx, ny): #takes into account # of iterations and z threshold, and number of x y integers
    x = np.linspace(-2, 2, nx) #allows me to define variables within function
    y = np.linspace(-2, 2, ny)
    c = x[:,np.newaxis] + 1j*y[np.newaxis,:] #newaxis allows for the dimension to be increased from 1D to 2D array for both x and y
    z=c
    for i in range(niter): #mandlebrot equation defined within this for loop as to iterate its results
         z = z**2 + c
    brotset = (abs(z) < thresh) #satisfies the condition

    return brotset
brotset = MandleBrot(50, 2, 300, 300) #defines brotset to specific parameters

plt.imshow(brotset.T, origin="lower", extent=[-2, 2, -2, 2], cmap="hot")
plt.show()   

plt.imshow(brotset.T, origin="lower", extent=[-2, 2, -2, 2], cmap="hot")
plt.gray()
plt.show()

            


```
